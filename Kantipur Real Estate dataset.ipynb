## Gathering the required Data

```python
# Title of each card
Prop_Name = []
Prop_Name_tags = soup_data.find_all('div', class_='image-title')
for image_title in Prop_Name_tags:
    Name = image_title.find('a').text.strip()
    Prop_Name.append(Name)
Prop_Name
```

```python
final_price = []
raw_price = soup_data.find_all('div',class_='property-price')
for price in raw_price:
    final_price.append(price.text.strip())
```

```python
final_price
```

```python
import re

cleaned_prices = [
    ' '.join(re.sub(r'\(.*?\)', '', p).split())
    for p in final_price
]

cleaned_prices
```

```python
## Converting text price to integer price

import re

def price_to_number(price_text):
    # Remove 'Rs.' and extra spaces
    price_text = price_text.replace('Rs.', '').strip()

    # Initialize total
    total = 0

    # Look for Crore
    crore_match = re.search(r'(\d+)\s*Crore', price_text)
    if crore_match:
        total += int(crore_match.group(1)) * 10_000_000  # 1 Crore = 10,000,000

    # Look for Lakh
    lakh_match = re.search(r'(\d+)\s*Lakh', price_text)
    if lakh_match:
        total += int(lakh_match.group(1)) * 100_000       # 1 Lakh = 100,000

    return total
```

```python
final_prices = [price_to_number(p) for p in cleaned_prices]
final_prices
```

# Extracting Data for all properties


## Slug Preparation


```python
def make_slug(Prop_Name):
    return (
        Prop_Name.lower().strip().replace(' ', '-').replace(',', '')
    )
df['slug'] = df['Title'].apply(make_slug)
df
```

```python
base_url = 'https://www.kantipurrealestate.com/property/'
df["base_url"] = base_url + df["slug"]
```

```python
df["base_url"].iloc[0]
```

```python
import requests
from bs4 import BeautifulSoup as Soup
import pandas as pd

all_property_details_list = []

total_properties = len(df)
print(f"Starting to scrape details for {total_properties} properties...")

for index, row in df.iterrows():
    url = row['base_url']

    if (index + 1) % 50 == 0 or (index + 1) == total_properties:
        print(f"Processing property {index + 1}/{total_properties}: {url}")

    property_data = {'base_url': url} # Use 'base_url' as the key to merge later

    try:
        response = requests.get(url, timeout=(5, 10)) # Added timeout
        response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

        single_data_soup = Soup(response.content, 'html.parser')

        # Extract Property Code directly from the property-detail-content block
        code_div = single_data_soup.find('div', class_='property-code')
        if code_div:
            code_span = code_div.find('span')
            if code_span:
                property_data['Property Code'] = code_span.get_text(strip=True) or None
        else:
            property_data['Property Code'] = None

        # Extract details from all 'other-details' sections (Area & Other Details, Accessibility, Room & Floor Details, Near By)
        other_details_divs = single_data_soup.find_all('div', class_='other-details')

        for details_div in other_details_divs:
            for li in details_div.find_all('li'):
                span = li.find('span')
                if span:
                    value = span.get_text(strip=True)
                    key_text = li.get_text(strip=True).replace(value, '').strip()

                    # Clean up the key text
                    if key_text.endswith(':'):
                        key = key_text[:-1].strip()
                    else:
                        key = key_text.strip()

                    if key:
                        property_data[key] = value if value else None

    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        property_data['Scrape_Error'] = str(e) # Log error
    except Exception as e:
        print(f"An unexpected error occurred for {url}: {e}")
        property_data['Scrape_Error'] = str(e)

    all_property_details_list.append(property_data)

# Create a DataFrame from the collected property details
detailed_properties_df = pd.DataFrame(all_property_details_list)

# Merge with the original df DataFrame
df_combined = pd.merge(df, detailed_properties_df, on='base_url', how='left')

# Update the main df DataFrame
df = df_combined

print("Scraping complete. The 'df' DataFrame has been updated with detailed property information.")
print(df.head())
print(df.info()) # Show column names
```
